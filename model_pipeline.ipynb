```python
from sentence_transformers import SentenceTransformer
```

```python
df2["combined_data"] = (
    df2["major"].fillna("")+" "+
    df2["skills"].fillna("")+" "+
    df2["interests"].fillna("")+" "+
    df2["bio"].fillna("")
).str.strip()
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
embeddings = model.encode(df2["combined_data"].tolist(), normalize_embeddings = True, show_progress_bar = True)
```

```python
np.save("embeddings.npy", embeddings)
print("Embeddings generated:", embeddings.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_profiles(user_id, df, embeddings, top_k=5, min_sim=0.7):
    try:
        user_idx = df[df["id"] == user_id].index[0]

        sim_scores = cosine_similarity([embeddings[user_idx]], embeddings).flatten()

        candidates = np.where(sim_scores > min_sim)[0]
        candidates = candidates[candidates != user_idx]  

        if len(candidates) == 0:
            return []

        top_idx = np.argsort(sim_scores[candidates])[::-1][:min(top_k, len(candidates))]

        results = df.iloc[candidates[top_idx]][
            ['id', 'major', 'interests', 'skills', 'is_alumni', 'is_mentor']
        ].copy()

        results['similarity_score'] = sim_scores[candidates[top_idx]]

        return results.to_dict('records')

    except IndexError:
        return {"error": f"User ID {user_id} not found"}
```

```python
recs = recommend_profiles(3,df2,embeddings)
```

```python
def recommend_mentors(student_id, df, embeddings, top_k = 5, min_sim = 0.7, year_diff_max = 5):
    try:
        student_row = df[df['id'] == student_id]
        if student_row.empty or student_row['is_alumni'].iloc[0]:
            return {"error": f"Student ID {student_id} not found or is not a student"}
        student_row = student_row.iloc[0]
        
        # Primary filter: alumni, mentors, within year_diff_max
        mentors_df = df[(df['is_alumni'] == True) & 
                        (df['is_mentor'] == True) & 
                        (abs(df['graduation_year'] - student_row['graduation_year']) <= year_diff_max)]
        
        # If no matches, relax year constraint
        if mentors_df.empty:
            mentors_df = df[(df['is_alumni'] == True) & (df['is_mentor'] == True)]
        
        # If still empty, relax mentor constraint but keep alumni
        # if mentors_df.empty:
        #     mentors_df = df[(df['is_alumni'] == True)]
        
        # If still empty, return error (should be rare with 5000 rows)
        if mentors_df.empty:
            return {"error": "No mentors available"}
        
        # Compute similarity
        mentor_emb = embeddings[mentors_df.index]
        student_emb = embeddings[student_row.name]
        sim_scores = cosine_similarity([student_emb], mentor_emb)[0]
        
        # Filter by similarity threshold
        valid_idx = np.where(sim_scores > min_sim)[0]
        if len(valid_idx) == 0:
            # Fallback: Return top-k mentors in same major, regardless of similarity
            mentors_df = mentors_df[mentors_df['major'] == student_row['major']]
            if mentors_df.empty:
                mentors_df = df[(df['is_alumni'] == True)]  # Ultimate fallback
            mentor_emb = embeddings[mentors_df.index]
            sim_scores = cosine_similarity([student_emb], mentor_emb)[0]
            valid_idx = np.arange(len(mentors_df))
        
        # Sort and select top-k
        top_idx = np.argsort(sim_scores[valid_idx])[::-1][:min(top_k, len(valid_idx))]
        results = mentors_df.iloc[valid_idx[top_idx]][['id', 'major', 'interests', 'skills', 'is_alumni', 'is_mentor']].copy()
        results['similarity_score'] = sim_scores[valid_idx[top_idx]]
        return results.to_dict('records')
    except IndexError:
        return {"error": f"Student ID {student_id} not found"}
    
```

```python
rec1 = recommend_mentors(10,df2,embeddings)
print(rec1)
```
